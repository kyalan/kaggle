# Pre-process Data

```python
def preprocessor_0(df_raw):

    df = df_raw.copy()
    
    df.drop(['Id'], axis=True, inplace=True)
    df.MSSubClass = df.MSSubClass.astype('str')
    
    return df
```

```python
df = preprocessor_0(df_raw)
```

### Train-dev Split

```python
from sklearn.model_selection import train_test_split

df_X = df.drop('SalePrice', axis=True)
df_y = df['SalePrice']

X_train, X_dev, y_train, y_dev = train_test_split(df_X, df_y, test_size=0.1, random_state=9527)
print('df is split. Train X, y shapes = {}, {}. Dev X, y shapes = {}, {}'.format(X_train.shape, y_train.shape, X_dev.shape, y_dev.shape))
```

```python
pd.set_option('display.max_columns', None)
df_X.head()
```

```python
numeric_features = [col for col in X_train if X_train[col].dtype in ['int64', 'float64']]
categorical_features = [col for col in X_train if not X_train[col].dtype in ['int64', 'float64']]

print('columns no. {} = num cols {} + cat cols {}'. format(len(X_train.columns), len(numeric_features), len(categorical_features)))
print('numeric_features: ', numeric_features)
print('categorical_features: ', categorical_features)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
    , ('logtransformer', FunctionTransformer(np.log1p))
    , ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor_1 = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
X_train = preprocessor_1.fit_transform(X_train)
X_dev = preprocessor_1.transform(X_dev)

print('X is transformed by sklearn preprocessor. Train X, y shapes = {}, {}. Dev X, y shapes = {}, {}'.format(X_train.shape, y_train.shape, X_dev.shape, y_dev.shape))
```

# Model Lab

```python
models = []

from xgboost import XGBRegressor

param = {
        'n_estimators': 100, 'learning_rate': 0.1, 'reg_lambda': 5
#         'objective':'reg:squaredlogerror', 'n_estimators': 100, 'n_jobs': 4
#          , 'max_depth': 50, 'learning_rate': 0.3, 'reg_lambda': 2
#          , 'gamma': 2, 'max_delta_step': 1, 'min_child_weight': 1
#          , 'colsample_bytree': 0.65, 'subsample': 0.9, 'base_score': 0.5
        }
models.append(XGBRegressor(**param))

from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
param = {
#     'n_estimators': 180, 'min_samples_split': 3
#          , 'min_samples_leaf': 4, 'max_depth': 32, 'bootstrap': True
#          , 'max_features': 0.5
        }
models.append(RandomForestRegressor(**param))
models.append(AdaBoostRegressor())
models.append(GradientBoostingRegressor())

from sklearn.tree import DecisionTreeRegressor
models.append(DecisionTreeRegressor())

from sklearn.neighbors import KNeighborsRegressor
models.append(KNeighborsRegressor())

from sklearn.svm import SVR
models.append(SVR())

print('Models list imported. with length = {}'.format(len(models)))
```

```python
from sklearn.metrics import mean_squared_log_error
df_acc = pd.DataFrame()

for model in models:
    model_name = model.__class__.__name__
    print('Running {}...'.format(model_name))
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    rmsle_train = np.sqrt(mean_squared_log_error(y_train, y_train_pred))
    y_dev_pred = model.predict(X_dev)
    rmsle_dev = np.sqrt(mean_squared_log_error(y_dev, y_dev_pred))
    
    df_acc = df_acc.append({'name': model_name, 'rmsle_dev': rmsle_dev, 'rmsle_train': rmsle_train}, ignore_index=True)
        
print(df_acc)
```

### Fine-tuning Model

```python
from sklearn.model_selection import GridSearchCV

grid = {'n_estimators':[100]
        , 'learning_rate':[0.1]
        , 'reg_alpha':[2]
        , 'reg_lambda':[4]
        , 'max_depth':[5]
        , 'objective':['reg:tweedie']
       }

search = GridSearchCV(estimator=XGBRegressor(), param_grid=grid, scoring='neg_mean_squared_log_error', n_jobs=4, refit=True)
search.fit(X_train, y_train)

print(search.best_params_)
print(search.best_score_)
```

# Modelling

```python
def getModel():
    from xgboost import XGBRegressor
    
    param = {'n_estimators':1000, 'learning_rate':0.1, 'reg_alpha':2
        , 'reg_lambda':4, 'max_depth':5, 'objective':'reg:tweedie', 'verbosity':0}
    model = XGBRegressor(**param)
    
    return model

model = getModel()
```

### Model Conclusion

```python
model.fit(X_train, y_train
          , eval_set=[(X_train, y_train), (X_dev, y_dev)]
          , eval_metric=['rmsle'], verbose=True
         )
```

```python
from matplotlib import pyplot

# retrieve performance metrics
results = model.evals_result()
epochs = len(results['validation_0']['rmsle'])-100
x_axis = range(0, epochs)
# plot log loss
fig, ax = pyplot.subplots()
ax.plot(x_axis, results['validation_0']['rmsle'][100:], label='Train')
ax.plot(x_axis, results['validation_1']['rmsle'][100:], label='Test')
ax.legend()
pyplot.show()
```
